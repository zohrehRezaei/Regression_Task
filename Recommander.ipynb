```python
import pandas as pd
import numpy as np

df=pd.read_csv("kc_house_data.csv")
pd.set_option('display.max_columns', 500)
print("Let me help you. There are 21613 cases in the range of 75000 to 7700000 $")
print("How much money do you have?")
while True:
    try:
        p=int(input())
    except ValueError:
        print("Invalid input! Please Enter a Number (75000 to 7700000)")
        continue
    if p>7700000 or p<75000:
        print("Invalid input! Please Enter a Number (75000 to 77000000)")
        continue
    else:
        df1=df.loc[df['price']<p]
        print("Number of cases for you:",len(df1))
        print("Top 10 cases based on your budget: ")
        df1=df1.sort_values(by="price", ascending=False)
        display(df1.head(10))
        print("How many rooms do you need? There are cases with 0 to 8 Rooms ")
        while True:
            try:
                b=int(input())
            except ValueError:
                print("Invalid input! Please Enter a Number (0 to 33)")
                continue
            if b>33 or b<0:
                print("Invalid input! Please Enter a Number (0 to 33)")
                continue
            else:
                df2=df1.loc[df['bedrooms']==b]
                print("Number of cases for you:",len(df2))
                print("Top 10 cases suitable for you: ")
                df2=df2.sort_values(by="price", ascending=False)
                display(df2.head(10))
                print("How many rooms do you need? There are cases with 0 to 8 Rooms ")
            
            
                
```
